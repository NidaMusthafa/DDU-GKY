## Single class single column

### Count of passanger survived from each class

```python
Pclass_survived=data.groupby(['Pclass'])['Survived']
```

```python
#print(Pclass_survived)
```

```python
Pclass_survived.sum()
```

## Multiple column

### Total amount paid by each passenger who survived

```python
Pclass_survived_fare=data.groupby(['Pclass'])['Survived','Fare']
```

```python
Pclass_survived_fare.sum()
```

## Multiple groups

### Count of males and females who survived from each class

```python
Pclass_sex_survived=data.groupby(['Pclass','Sex'],as_index=False)['Survived']
```

```python
Pclass_sex_survived.sum()
```

## Multiple funcion

### Mean and Sum of pclass survived

```python
pclass_Survived_mean_sum=data.groupby(['Pclass'])['Survived'].agg(['mean','sum'])
```

```python
pclass_Survived_mean_sum.sum()
```
