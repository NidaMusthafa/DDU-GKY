## Handling Null Values

```python
data=pd.read_csv('Windspeed.csv')
data.head()
```

```python
data.isnull().sum()
```

### Filling a common value to all missing data

```python
data.fillna(0)
```

## Adding missing data to individual columns

```python
data.fillna({
    "Temperature":0,
    "Windspeed":5,
    "Status":"sunny"
})
```

## Forward fill

```python
data.fillna(method="ffill")
```

## Backward fill

```python
data.fillna(method='bfill')
```

## Limiting the forward/backward fill

```python
data.fillna(method='ffill',limit=1)
```

## Filling with pandas obj

```python
data.fillna(data.mean())
```

## Filling for specific range of columns

```python
data.fillna(data.mean()['Temperature':'Windspeed'])
```

## Interpolate missing value

```python
data.interpolate()
```

## Time interpolate

```python
data.interpolate(limit=1,limit_direction='backward')
```

## Limit area interpolation

```python
data.isnull().sum()
```
